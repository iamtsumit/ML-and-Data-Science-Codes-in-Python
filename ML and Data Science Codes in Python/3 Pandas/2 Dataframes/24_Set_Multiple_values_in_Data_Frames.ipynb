```
bond = pd.read_csv('jamesbond.csv', index_col='Film')
```

```
bond.sort_index(inplace=True)
```

```
mask = bond['Actor'] == 'Sean Connery'
```

```
bond[mask]
```

```
bond[mask]['Actor'] = 'Sir Sean Connery' # give warning as it doesnot make changes in actual data it only makes changes in piece of data frame
```

```
bond[mask].head()
```

```
df2 = bond[mask]
```

```
df2['Actor'] = 'Sir Sean Connery' #still does not make changes to actual data
```

```
df2.head()
```

```
bond.loc[mask]
```

```
bond.loc[mask,'Actor'] = 'Sir Sean Connery' #make changes in actual data
```
