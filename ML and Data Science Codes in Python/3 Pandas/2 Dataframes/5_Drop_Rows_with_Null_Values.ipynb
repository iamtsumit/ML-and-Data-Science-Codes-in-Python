```
nba = pd.read_csv('nba.csv')
```

```
nba.head()
```

```
nba.dropna() #Remove row even one column contains NAN
```

```
nba.dropna(how = 'any') #Remove row even one column contains NAN
```

```
nba.dropna(how = 'all') #If all attribute of row are NAN is removed
```

```
nba.dropna(how = 'all', inplace=True) #new dataset to replace previous dataset
```

```
nba.tail()
```

```
nba.dropna(axis = 1) #The column contains NA are removed 
```

```
nba.dropna(subset=['Salary']) #The row which contains NAN in Salary column is removed
```
