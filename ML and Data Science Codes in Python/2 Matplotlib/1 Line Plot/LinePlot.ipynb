```
t = np.arange(0.0,2.0,0.1) #time
```

```
t
```

```
s = 1 + np.cos(2*np.pi*t) #cos
```

```
s
```

```
plt.plot(t,s)
plt.xlabel('Time (t)')
plt.ylabel('Voltage (mV)')
plt.title('Cosine Wave')
plt.show()
```

```
plt.plot(t,s,'g')
plt.xlabel('Time (t)')
plt.ylabel('Voltage (mV)')
plt.title('Cosine Wave')
plt.show()
```

```
plt.plot(t,s,'*')
plt.xlabel('Time (t)')
plt.ylabel('Voltage (mV)')
plt.title('Cosine Wave')
plt.show()
```

```
plt.plot(t,s,'--')
plt.grid()
plt.xlabel('Time (t)')
plt.ylabel('Voltage (mV)')
plt.title('Cosine Wave')
plt.show()
```
